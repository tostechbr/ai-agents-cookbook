```python
from typing import TypedDict
from langgraph.graph import StateGraph, END

# 1. Define your State
class MyState(TypedDict):
    message: str
    count: int

print("State defined!")
```

### Exercise 1: Create Nodes

Create two functions: `node_a` and `node_b`. 
- `node_a` should add " Hello" to the message.
- `node_b` should add " World" to the message.

```python
# Your code here
def node_a(state: MyState):
    return {"message": state['message'] + " Hello"}

def node_b(state: MyState):
    return {"message": state['message'] + " World"}
```

### Exercise 2: Build the Graph

Now, let's build the graph!
1. Initialize `StateGraph` with `MyState`.
2. Add `node_a` and `node_b`.
3. Set the entry point to `node_a`.
4. Add an edge from `node_a` to `node_b`.
5. Add an edge from `node_b` to `END`.
6. Compile the graph.

```python
# Your code here
builder = StateGraph(MyState)

# Add nodes
builder.add_node("node_a", node_a)
builder.add_node("node_b", node_b)

# Define flow
builder.set_entry_point("node_a")
builder.add_edge("node_a", "node_b")
builder.add_edge("node_b", END)

# Compile
graph = builder.compile()
print("Graph compiled!")
```

### Exercise 3: Run the Graph

Invoke the graph with an initial state.
Initialize `message` as `"Start:"` and `count` as `0`.

```python
# Your code here
result = graph.invoke({"message": "Start:", "count": 0})
print(result)
```
